```python
import xarray as xr
import os 
scratch = os.environ['SCRATCH']
```

```python
from datetime import datetime
datetime.now()
```

```python
def add_ensemble_dim(xda):
    xda = xda.expand_dims(var = [datetime.now()])
    return xda
```

```python
datas = []
for ensemble_mem in range(1,11):
    data = xr.open_dataset(f'{scratch}/1960_{ensemble_mem}.nc')
    datas.append(data['gpp'].mean(axis=(1,2)))
import numpy as np
ds = np.stack(datas)
import matplotlib.pyplot as plt 

for i in range(10):
    plt.plot(ds[i,:])
```
