```python
experiments = ['flow_no_climatology_no_delta',
'flow_sub_pixel_no_delta_no_smoothing',
'flow_conv_head_layernorm',
'ddpm_elevation_scaled_1000_timesteps']
```

```python
import glob
import os 
import torch
work = os.environ['WORK']
```

```python
metrics_output = []
for exp in experiments:
    metrics_output_1 = []
    metrics = torch.load(f'{work}/ipsl_dcpp/ipsl_dcpp/{exp}_sharpnesses.pt')
    for i in metrics:
        metrics_output_1.append(torch.stack(i))
    metrics_output.append(torch.stack(metrics_output_1))
metrics_output = torch.stack(metrics_output)
```

```python
print(metrics_output[0,0,:,9])
```

```python
import matplotlib.pyplot as plt

for i in range(4):
    plt.plot(metrics_output[i,0,:,4],label=experiments[i])
plt.legend()
```
